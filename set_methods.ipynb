# Set Built-in Methods in Python

This notebook covers all the important built-in methods available for sets in Python, with examples and explanations.

## Basic Set Operations

### 1. copy()
Performs a shallow copy of the set.

**Example:**
```python
s = {1, 2, 3}
s2 = s.copy()
```

````python
# Example: copy()
s = {1, 2, 3, 4, 5}
print(f"Original set: {s}")

s2 = s.copy()
print(f"Copied set: {s2}")

# Modify original to show they're independent
s.add(6)
print(f"After adding 6 to original: {s}")
print(f"Copied set remains: {s2}")
````

### 2. clear()
Deletes all elements from the set but doesn't delete the memory. The set becomes empty.

**Example:**
```python
s = {2, 5, 10, 1}
s.clear()  # s becomes {}
```

````python
# Example: clear()
s = {2, 5, 10, 1}
print(f"Before clear: {s}")

s.clear()
print(f"After clear: {s}")
print(f"Type after clear: {type(s)}")  # Still a set, just empty
````

### 3. add(value)
Adds an element to the set. Since sets are unordered, the element is added randomly.

**Example:**
```python
s = {3, 2, 8}
s.add(12)  # s becomes {2, 12, 8, 3} (order may vary)
```

````python
# Example: add()
s = {3, 2, 8}
print(f"Original set: {s}")

s.add(12)
print(f"After adding 12: {s}")

s.add(12)  # Adding duplicate element
print(f"After adding 12 again: {s}")  # No change, sets don't allow duplicates
````

## Removing Elements from Sets

### 4. pop()
Removes and returns the first element from the set. Takes no arguments. If the set is empty, it raises a KeyError.

**Example:**
```python
s = {3}
s.pop()  # Returns 3, set becomes empty
s.pop()  # KeyError
```

````python
# Example: pop()
s = {10, 20, 30, 40}
print(f"Original set: {s}")

popped_element = s.pop()
print(f"popped element: {popped_element}")
print(f"After pop: {s}")

# Uncomment to see KeyError on empty set
# s.clear()
# s.pop()
````

### 5. remove(value)
Removes the specified element from the set. If the element is not present, it raises a KeyError. Does not return the deleted element.

**Example:**
```python
s = {2, 12, 8, 3}
s.remove(8)  # s becomes {2, 12, 3}
s.remove(15)  # KeyError
```

````python
# Example: remove()
s = {2, 12, 8, 3}
print(f"Original set: {s}")

s.remove(8)
print(f"After remove(8): {s}")

# Uncomment to see KeyError
# s.remove(15)
````

### 6. discard(value)
Removes the specified element from the set if present. If the element is not present, it does nothing (no error). Does not return the deleted element.

**Example:**
```python
s = {2, 12, 8, 3}
s.discard(8)  # s becomes {2, 12, 3}
s.discard(15)  # No error, set remains unchanged
```

````python
# Example: discard()
s = {2, 12, 8, 3}
print(f"Original set: {s}")

s.discard(8)
print(f"After discard(8): {s}")

s.discard(15)  # Element doesn't exist
print(f"After discard(15): {s}")  # No change, no error
````

## Set Operations (Return New Sets)

### 7. union()
Returns all elements from both the base set and the specified set(s). Does not modify the original sets.

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.union(s2)  # Returns {1, 2, 3, 7}
```

````python
# Example: union()
s = {1, 2, 3}
s2 = {1, 2, 7}
s3 = {5, 6}

print(f"Set s: {s}")
print(f"Set s2: {s2}")

result = s.union(s2)
print(f"s.union(s2): {result}")

result = s.union(s2, s3)
print(f"s.union(s2, s3): {result}")

# Original sets unchanged
print(f"Original s still: {s}")
print(f"Original s2 still: {s2}")
````

### 8. intersection()
Returns common elements from both sets. Does not modify the original sets.

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.intersection(s2)  # Returns {1, 2}
```

````python
# Example: intersection()
s = {1, 2, 3, 4}
s2 = {1, 2, 7, 8}
s3 = {1, 9, 10}

print(f"Set s: {s}")
print(f"Set s2: {s2}")

result = s.intersection(s2)
print(f"s.intersection(s2): {result}")

result = s.intersection(s2, s3)
print(f"s.intersection(s2, s3): {result}")
````

### 9. difference()
Returns elements present in the base set but not in the specified set(s). Does not modify the original sets.

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.difference(s2)  # Returns {3}
```

````python
# Example: difference()
s = {1, 2, 3, 4, 5}
s2 = {1, 2, 7}
s3 = {3, 4}

print(f"Set s: {s}")
print(f"Set s2: {s2}")

result = s.difference(s2)
print(f"s.difference(s2): {result}")

result = s.difference(s2, s3)
print(f"s.difference(s2, s3): {result}")
````

### 10. symmetric_difference()
Returns elements that are in either set, but not in both (uncommon elements from both sets). Does not modify the original sets.

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.symmetric_difference(s2)  # Returns {3, 7}
```

````python
# Example: symmetric_difference()
s = {1, 2, 3, 4}
s2 = {1, 2, 7, 8}

print(f"Set s: {s}")
print(f"Set s2: {s2}")

result = s.symmetric_difference(s2)
print(f"s.symmetric_difference(s2): {result}")
````

## Set Operations (Modify Original Set)

> **Note:** These methods modify the base set and do not return a new set.

### 11. update()
Updates the base set by adding all elements from the specified set(s).

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.update(s2)  # s becomes {1, 2, 3, 7}
```

````python
# Example: update()
s = {1, 2, 3}
s2 = {1, 2, 7}
s3 = {8, 9}

print(f"Original s: {s}")
print(f"s2: {s2}")

s.update(s2)
print(f"After s.update(s2): {s}")

s.update(s3)
print(f"After s.update(s3): {s}")
````

### 12. intersection_update()
Updates the base set by keeping only elements found in both sets (intersection).

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.intersection_update(s2)  # s becomes {1, 2}
```

````python
# Example: intersection_update()
s = {1, 2, 3, 4}
s2 = {1, 2, 7, 8}

print(f"Original s: {s}")
print(f"s2: {s2}")

s.intersection_update(s2)
print(f"After s.intersection_update(s2): {s}")
````

### 13. difference_update()
Updates the base set by removing elements found in the specified set(s).

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.difference_update(s2)  # s becomes {3}
```

````python
# Example: difference_update()
s = {1, 2, 3, 4, 5}
s2 = {1, 2, 7}

print(f"Original s: {s}")
print(f"s2: {s2}")

s.difference_update(s2)
print(f"After s.difference_update(s2): {s}")
````

### 14. symmetric_difference_update()
Updates the base set by keeping only elements that are in either set, but not in both.

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2, 7}
s.symmetric_difference_update(s2)  # s becomes {3, 7}
```

````python
# Example: symmetric_difference_update()
s = {1, 2, 3, 4}
s2 = {1, 2, 7, 8}

print(f"Original s: {s}")
print(f"s2: {s2}")

s.symmetric_difference_update(s2)
print(f"After s.symmetric_difference_update(s2): {s}")
````

## Set Comparison Methods

### 15. issuperset()
Returns True if all elements of the specified set are present in the base set, otherwise False.

**Example:**
```python
s = {1, 2, 3}
s2 = {1, 2}
s.issuperset(s2)  # Returns True
```

````python
# Example: issuperset()
s = {1, 2, 3, 4, 5}
s2 = {1, 2, 3}
s3 = {1, 2, 6}

print(f"Set s: {s}")
print(f"Set s2: {s2}")
print(f"Set s3: {s3}")

print(f"s.issuperset(s2): {s.issuperset(s2)}")
print(f"s.issuperset(s3): {s.issuperset(s3)}")
````

### 16. issubset()
Returns True if all elements of the base set are present in the specified set, otherwise False.

**Example:**
```python
s = {1, 2}
s2 = {1, 2, 3}
s.issubset(s2)  # Returns True
```

````python
# Example: issubset()
s = {1, 2, 3}
s2 = {1, 2, 3, 4, 5}
s3 = {1, 2, 6}

print(f"Set s: {s}")
print(f"Set s2: {s2}")
print(f"Set s3: {s3}")

print(f"s.issubset(s2): {s.issubset(s2)}")
print(f"s.issubset(s3): {s.issubset(s3)}")
````

### 17. isdisjoint()
Returns True if the sets have no common elements, otherwise False.

**Example:**
```python
s = {8, 9}
s2 = {1, 2, 3}
s.isdisjoint(s2)  # Returns True
```

````python
# Example: isdisjoint()
s = {8, 9, 10}
s2 = {1, 2, 3}
s3 = {1, 8, 9}

print(f"Set s: {s}")
print(f"Set s2: {s2}")
print(f"Set s3: {s3}")

print(f"s.isdisjoint(s2): {s.isdisjoint(s2)}")
print(f"s.isdisjoint(s3): {s.isdisjoint(s3)}")
````

## Summary Table

| Method | Description | Returns New Set? | Modifies Original? |
|--------|-------------|-----------------|-------------------|
| `copy()` | Shallow copy | Yes (new set) | No |
| `clear()` | Remove all elements | No | Yes |
| `add(x)` | Add element | No | Yes |
| `pop()` | Remove and return arbitrary element | No | Yes |
| `remove(x)` | Remove specific element (error if missing) | No | Yes |
| `discard(x)` | Remove specific element (no error if missing) | No | Yes |
| `union()` | All elements from both sets | Yes | No |
| `intersection()` | Common elements | Yes | No |
| `difference()` | Elements in base set only | Yes | No |
| `symmetric_difference()` | Elements in either set, but not both | Yes | No |
| `update()` | Add all elements from another set | No | Yes |
| `intersection_update()` | Keep only common elements | No | Yes |
| `difference_update()` | Remove elements found in another set | No | Yes |
| `symmetric_difference_update()` | Keep elements in either set, but not both | No | Yes |
| `issuperset()` | Check if set contains another set | Boolean | No |
| `issubset()` | Check if set is contained in another set | Boolean | No |
| `isdisjoint()` | Check if sets have no common elements | Boolean | No |

## Visual Examples of Set Operations

````python
# Complete example showing all set operations
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(f"Set A: {A}")
print(f"Set B: {B}")
print("-" * 30)

# Operations that return new sets
print(f"A ∪ B (union): {A.union(B)}")
print(f"A ∩ B (intersection): {A.intersection(B)}")
print(f"A - B (difference): {A.difference(B)}")
print(f"B - A (difference): {B.difference(A)}")
print(f"A △ B (symmetric difference): {A.symmetric_difference(B)}")
print("-" * 30)

# Comparison operations
print(f"A is superset of B: {A.issuperset(B)}")
print(f"A is subset of B: {A.issubset(B)}")
print(f"A and B are disjoint: {A.isdisjoint(B)}")
````
