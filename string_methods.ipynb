# String Built-in Methods in Python

This notebook covers all the important built-in methods available for strings in Python, with examples and explanations.

## Case Conversion Methods

### 1. lower()
Converts the given string into lowercase.

**Example:**
```python
s = 'HeLlO'
s.lower()  # Output: 'hello'
```

````python
# Example: lower()
s = 'HeLlO'
print(f"Original string: {s}")
print(f"After lower(): {s.lower()}")
````

### 2. upper()
Converts the given string into uppercase.

**Example:**
```python
s = 'HeLlO'
s.upper()  # Output: 'HELLO'
```

````python
# Example: upper()
s = 'HeLlO'
print(f"Original string: {s}")
print(f"After upper(): {s.upper()}")
````

### 3. title()
Converts the first character of each word to uppercase and the rest to lowercase.

**Example:**
```python
s = 'hElLo pythoN'
s.title()  # Output: 'Hello Python'
```

````python
# Example: title()
s = 'hElLo pythoN'
print(f"Original string: {s}")
print(f"After title(): {s.title()}")
````

### 4. capitalize()
Converts the first character of the entire string to uppercase and the rest to lowercase.

**Example:**
```python
s = 'hElLo pythoN'
s.capitalize()  # Output: 'Hello python'
```

````python
# Example: capitalize()
s = 'hElLo pythoN'
print(f"Original string: {s}")
print(f"After capitalize(): {s.capitalize()}")
````

### 5. swapcase()
Converts uppercase characters to lowercase and lowercase characters to uppercase.

**Example:**
```python
s = 'HeLlO PYTHoN'
s.swapcase()  # Output: 'hElLo python'
```

````python
# Example: swapcase()
s = 'HeLlO PYTHoN'
print(f"Original string: {s}")
print(f"After swapcase(): {s.swapcase()}")
````

## Boolean Check Methods (is* methods)

> **Note:** If any method starts with 'is', it returns a Boolean value (True/False).

### 6. islower()
Returns True if the string is in lowercase, otherwise False.

**Example:**
```python
s = 'hello'
s.islower()  # Output: True
```

````python
# Example: islower()
s1 = 'hello'
s2 = 'Hello'
print(f"'{s1}'.islower(): {s1.islower()}")
print(f"'{s2}'.islower(): {s2.islower()}")
````

### 7. isupper()
Returns True if the string is in uppercase, otherwise False.

**Example:**
```python
s = 'HELLO'
s.isupper()  # Output: True
```

````python
# Example: isupper()
s1 = 'HELLO'
s2 = 'Hello'
print(f"'{s1}'.isupper(): {s1.isupper()}")
print(f"'{s2}'.isupper(): {s2.isupper()}")
````

### 8. istitle()
Returns True if the string is in title case, otherwise False.

**Example:**
```python
s = 'Hello'
s.istitle()  # Output: True
```

````python
# Example: istitle()
s1 = 'Hello Python'
s2 = 'Hello python'
print(f"'{s1}'.istitle(): {s1.istitle()}")
print(f"'{s2}'.istitle(): {s2.istitle()}")
````

### 9. isalpha()
Returns True if the string contains only alphabets, otherwise False.

**Example:**
```python
s = 'hello'
s.isalpha()  # Output: True
```

````python
# Example: isalpha()
s1 = 'hello'
s2 = 'hello123'
print(f"'{s1}'.isalpha(): {s1.isalpha()}")
print(f"'{s2}'.isalpha(): {s2.isalpha()}")
````

### 10. isdigit()
Returns True if the string contains only digits, otherwise False.

**Example:**
```python
s = '1234'
s.isdigit()  # Output: True
```

````python
# Example: isdigit()
s1 = '1234'
s2 = '1234abc'
print(f"'{s1}'.isdigit(): {s1.isdigit()}")
print(f"'{s2}'.isdigit(): {s2.isdigit()}")
````

### 11. isalnum()
Returns True if the string contains only alphabets or digits, otherwise False.

**Example:**
```python
s = 'hello123'
s.isalnum()  # Output: True
```

````python
# Example: isalnum()
s1 = 'hello123'
s2 = 'hello 123'
print(f"'{s1}'.isalnum(): {s1.isalnum()}")
print(f"'{s2}'.isalnum(): {s2.isalnum()}")
````

### 12. isspace()
Returns True if the string contains only spaces, otherwise False.

**Example:**
```python
s = '  '
s.isspace()  # Output: True
```

````python
# Example: isspace()
s1 = '  '
s2 = '  a  '
print(f"'{s1}'.isspace(): {s1.isspace()}")
print(f"'{s2}'.isspace(): {s2.isspace()}")
````

## Search and Check Methods

### 13. startswith(substr, [start], [end])
Returns True if the string starts with the specified substring at the given index position.

**Example:**
```python
s = 'hello'
s.startswith('h')  # Output: True
```

````python
# Example: startswith()
s = 'hello'
print(f"'{s}'.startswith('h'): {s.startswith('h')}")
print(f"'{s}'.startswith('e'): {s.startswith('e')}")
print(f"'{s}'.startswith('e', 1, 3): {s.startswith('e', 1, 3)}")
````

### 14. endswith(substr, [start], [end])
Returns True if the string ends with the specified substring at the given index position.

**Example:**
```python
s = 'hello'
s.endswith('o')  # Output: True
s.endswith('l', 0, 2)  # Output: True
```

````python
# Example: endswith()
s = 'hello'
print(f"'{s}'.endswith('o'): {s.endswith('o')}")
print(f"'{s}'.endswith('l'): {s.endswith('l')}")
print(f"'{s}'.endswith('l', 0, 2): {s.endswith('l', 0, 2)}")
````

## Stripping Methods

### 15. strip([unwanted_char])
Removes unwanted characters (default is space) from both sides of the string.

**Examples:**
```python
s = '    hello    '
s.strip()  # Output: 'hello'

s = '### hello ###'
s.strip('#')  # Output: ' hello '
```

````python
# Example: strip()
s1 = '    hello    '
s2 = '### hello ###'
print(f"Original: '{s1}'")
print(f"After strip(): '{s1.strip()}'")
print(f"Original: '{s2}'")
print(f"After strip('#'): '{s2.strip('#')}'")
````

### 16. rstrip([unwanted_char])
Removes unwanted characters (default is space) from the right side only.

**Examples:**
```python
s = 'hello    '
s.rstrip()  # Output: 'hello'

s = '### hello ###'
s.rstrip('#')  # Output: '###hello'
```

````python
# Example: rstrip()
s1 = 'hello    '
s2 = '### hello ###'
print(f"Original: '{s1}'")
print(f"After rstrip(): '{s1.rstrip()}'")
print(f"Original: '{s2}'")
print(f"After rstrip('#'): '{s2.rstrip('#')}'")
````

### 17. lstrip([unwanted_char])
Removes unwanted characters (default is space) from the left side only.

**Examples:**
```python
s = '    hello'
s.lstrip()  # Output: 'hello'

s = '### hello ###'
s.lstrip('#')  # Output: 'hello###'
```

````python
# Example: lstrip()
s1 = '    hello'
s2 = '### hello ###'
print(f"Original: '{s1}'")
print(f"After lstrip(): '{s1.lstrip()}'")
print(f"Original: '{s2}'")
print(f"After lstrip('#'): '{s2.lstrip('#')}'")
````

## Splitting Methods

### 18. split([delimiter], [count])
Splits the string based on the given delimiter (default is space) and returns a list.

**Examples:**
```python
s = 'hello python'
s.split()  # Output: ['hello', 'python']

s = 'hello,python'
s.split(',')  # Output: ['hello', 'python']
```

````python
# Example: split()
s1 = 'hello python'
s2 = 'hello,python'
s3 = 'hello,python,world,programming'
print(f"'{s1}'.split(): {s1.split()}")
print(f"'{s2}'.split(','): {s2.split(',')}")
print(f"'{s3}'.split(',', 2): {s3.split(',', 2)}")
````

### 19. rsplit([delimiter], [count])
Splits the string from the right side based on the given delimiter. Splitting is done left to right but searching is from right to left.

**Examples:**
```python
s = 'hello python'
s.rsplit()  # Output: ['hello', 'python']

s = 'hello,python'
s.rsplit(',')  # Output: ['hello', 'python']
```

````python
# Example: rsplit()
s = 'hello,python,world,programming'
print(f"'{s}'.rsplit(',', 1): {s.rsplit(',', 1)}")
print(f"'{s}'.rsplit(',', 2): {s.rsplit(',', 2)}")
print(f"Compare with split(',', 2): {s.split(',', 2)}")
````

## Search and Replace Methods

### 20. replace(oldsubstr, newsubstr, [count])
Replaces all occurrences of old substring with new substring. Count limits the number of replacements.

**Examples:**
```python
s = 'hello python'
s.replace('l', '@')  # Output: 'he@@o python'
s.replace('l', '@', 1)  # Output: 'he@lo python'
```

````python
# Example: replace()
s = 'hello python hello world'
print(f"Original: '{s}'")
print(f"replace('l', '@'): {s.replace('l', '@')}")
print(f"replace('l', '@', 2): {s.replace('l', '@', 2)}")
print(f"replace('hello', 'hi'): {s.replace('hello', 'hi')}")
````

### 21. count(substr, [start], [end])
Returns the number of occurrences of the substring in the string.

**Examples:**
```python
s = 'hello python hai'
s.count('h')  # Output: 3
s.count('h', 0, 3)  # Output: 1
```

````python
# Example: count()
s = 'hello python hai'
print(f"'{s}'.count('h'): {s.count('h')}")
print(f"'{s}'.count('h', 0, 3): {s.count('h', 0, 3)}")
print(f"'{s}'.count('o'): {s.count('o')}")
````

### 22. index(substr, [start], [end])
Returns the first matched index position of the substring. Raises ValueError if not found.

**Examples:**
```python
s = 'hello'
s.index('l')  # Output: 2
s.index('z')  # ValueError
```

````python
# Example: index()
s = 'hello'
print(f"'{s}'.index('l'): {s.index('l')}")
print(f"'{s}'.index('l', 3): {s.index('l', 3)}")

# Uncomment to see ValueError
# print(s.index('z'))
````

### 23. find(substr, [start], [end])
Returns the first matched index position of the substring. Returns -1 if not found.

**Examples:**
```python
s = 'hello'
s.find('l')  # Output: 2
s.find('z')  # Output: -1
```

````python
# Example: find()
s = 'hello'
print(f"'{s}'.find('l'): {s.find('l')}")
print(f"'{s}'.find('z'): {s.find('z')}")
print(f"'{s}'.find('l', 3): {s.find('l', 3)}")
````

## Formatting and Joining Methods

### 24. format()
Represents static text with dynamic values. The modern way is using f-strings.

**Example:**
```python
name = 'John'
age = 25
f'My name is {name} and I am {age} years old'
```

````python
# Example: format() and f-strings
name = 'John'
age = 25

# Using f-string (modern approach)
print(f"My name is {name} and I am {age} years old")

# Using format() method
print("My name is {} and I am {} years old".format(name, age))

# Using format() with indices
print("My name is {0} and I am {1} years old".format(name, age))
````

### 25. join()
Joins the elements of a collection with the glue character and returns a string.

**Example:**
```python
s = 'hello'
'.'.join(s)  # Output: 'h.e.l.l.o'
```

````python
# Example: join()
s = 'hello'
words = ['hello', 'python', 'world']

print(f"'.'.join('{s}'): '{'.'.join(s)}'")
print(f"' '.join({words}): '{' '.join(words)}'")
print(f"'-'.join({words}): '{'-'.join(words)}'")
````

## Partition Methods

### 26. partition(delimiter)
Splits the string into three parts (head, delimiter, tail) and returns a tuple.

**Example:**
```python
s = 'hello python'
s.partition('o')  # Output: ('hell', 'o', ' python')
```

````python
# Example: partition()
s = 'hello python'
print(f"'{s}'.partition('o'): {s.partition('o')}")
print(f"'{s}'.partition(' '): {s.partition(' ')}")
print(f"'{s}'.partition('z'): {s.partition('z')}")
````

### 27. rpartition(delimiter)
Splits the string into three parts from the right. Searching is from right to left.

**Example:**
```python
s = 'hello python'
s.rpartition('o')  # Output: ('hello pyth', 'o', 'n')
```

````python
# Example: rpartition()
s = 'hello python'
print(f"'{s}'.rpartition('o'): {s.rpartition('o')}")
print(f"Compare with partition('o'): {s.partition('o')}")
print(f"'{s}'.rpartition(' '): {s.rpartition(' ')}")
````
