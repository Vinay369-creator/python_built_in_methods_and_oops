# Dictionary Built-in Methods in Python

This notebook covers all the important built-in methods available for dictionaries in Python, with examples and explanations.

## Basic Dictionary Operations

### 1. copy()
Performs a shallow copy of the dictionary.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d2 = d.copy()
```

````python
# Example: copy()
d = {'name': 'abc', 'age': 21, 'hobbies': ['reading', 'music']}
print(f"Original dictionary: {d}")

d2 = d.copy()
print(f"Copied dictionary: {d2}")

# Demonstrate shallow copy behavior
d['hobbies'].append('sports')
print("\nAfter modifying nested object in original:")
print(f"Original: {d}")
print(f"Copy: {d2}")  # Nested object changes affect both
````

### 2. clear()
Deletes all elements from the dictionary but doesn't delete the memory. The dictionary becomes empty.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d.clear()  # d becomes {}
```

````python
# Example: clear()
d = {'name': 'abc', 'age': 21}
print(f"Before clear: {d}")

d.clear()
print(f"After clear: {d}")
print(f"Type after clear: {type(d)}")  # Still a dictionary, just empty
````

## Retrieval Methods

### 3. keys()
Returns a view object containing all the keys in the dictionary.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d.keys()  # dict_keys(['name', 'age'])
```

````python
# Example: keys()
d = {'name': 'abc', 'age': 21, 'city': 'blr', 'mobile': 1234567890}
print(f"Dictionary: {d}")

keys = d.keys()
print(f"Keys: {keys}")
print(f"Type of keys: {type(keys)}")

# Convert to list if needed
print(f"Keys as list: {list(keys)}")

# Iterate through keys
print("\nIterating through keys:")
for key in d.keys():
    print(f"  {key}")
````

### 4. values()
Returns a view object containing all the values in the dictionary.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d.values()  # dict_values(['abc', 21])
```

````python
# Example: values()
d = {'name': 'abc', 'age': 21, 'city': 'blr', 'mobile': 1234567890}
print(f"Dictionary: {d}")

values = d.values()
print(f"Values: {values}")
print(f"Values as list: {list(values)}")

# Iterate through values
print("\nIterating through values:")
for value in d.values():
    print(f"  {value}")
````

### 5. items()
Returns a view object containing key-value pairs as tuples.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d.items()  # dict_items([('name', 'abc'), ('age', 21)])
```

````python
# Example: items()
d = {'name': 'abc', 'age': 21, 'city': 'blr'}
print(f"Dictionary: {d}")

items = d.items()
print(f"Items: {items}")
print(f"Items as list: {list(items)}")

# Iterate through items (most common way)
print("\nIterating through items:")
for key, value in d.items():
    print(f"  {key}: {value}")
````

## Safe Access Methods

### 6. get(keyname, [value])
Returns the value for the given key. If the key doesn't exist, it returns None (or the default value if provided).

**Examples:**
```python
d = {'name': 'abc', 'age': 21}
d.get('name')        # Returns 'abc'
d.get('mob')         # Returns None
d.get('mob', 1234)   # Returns 1234
```

````python
# Example: get()
d = {'name': 'abc', 'age': 21}
print(f"Dictionary: {d}")

print(f"get('name'): {d.get('name')}")
print(f"get('mob'): {d.get('mob')}")  # Returns None
print(f"get('mob', 1234): {d.get('mob', 1234)}")

# Compare with direct access
try:
    print(d['mob'])  # This would raise KeyError
except KeyError:
    print("d['mob'] raises KeyError - use get() to avoid this")

# Practical example
user_settings = {'theme': 'dark', 'notifications': True}
font_size = user_settings.get('font_size', 12)  # Provides default value
print(f"\nFont size setting: {font_size}")
````

### 7. setdefault(keyname, [value])
Returns the value for the given key. If the key doesn't exist, it inserts the key with the given value (or None) and returns that value.

**Examples:**
```python
d = {'name': 'abc', 'age': 21}
d.setdefault('mob')           # Adds 'mob': None
d.setdefault('add', 'blr')    # Adds 'add': 'blr'
```

````python
# Example: setdefault()
d = {'name': 'abc', 'age': 21}
print(f"Original dictionary: {d}")

# Key doesn't exist - adds with None
result = d.setdefault('mob')
print(f"setdefault('mob') returned: {result}")
print(f"After setdefault('mob'): {d}")

# Key doesn't exist - adds with specified value
result = d.setdefault('add', 'blr')
print(f"\nsetdefault('add', 'blr') returned: {result}")
print(f"After setdefault('add', 'blr'): {d}")

# Key exists - returns existing value, no change
result = d.setdefault('name', 'xyz')
print(f"\nsetdefault('name', 'xyz') returned: {result}")
print(f"After setdefault on existing key: {d}")  # 'name' unchanged

# Practical example - counting words
text = "apple banana apple orange apple"
word_count = {}
for word in text.split():
    word_count[word] = word_count.setdefault(word, 0) + 1
print(f"\nWord count using setdefault: {word_count}")
````

## Update Methods

### 8. update(key value pairs)
Updates the dictionary with key-value pairs from another dictionary or iterable.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d2 = {'add': 'blr'}
d.update(d2)  # d becomes {'name': 'abc', 'age': 21, 'add': 'blr'}
```

````python
# Example: update()
d = {'name': 'abc', 'age': 21}
print(f"Original dictionary: {d}")

# Update with another dictionary
d2 = {'add': 'blr', 'phone': 1234}
d.update(d2)
print(f"After update with d2: {d}")

# Update with key-value pairs (overwrites existing keys)
d.update({'city': 'mumbai', 'age': 22})  # Updates age from 21 to 22
print(f"After another update: {d}")

# Update with list of tuples
d.update([('country', 'india'), ('pin', 560001)])
print(f"After update with list of tuples: {d}")

# Update with keyword arguments
d.update(state='karnataka', email='abc@example.com')
print(f"After update with keyword args: {d}")
````

## Deletion Methods

### 9. pop(keyname)
Removes and returns the value for the given key. If the key doesn't exist, it raises a KeyError.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d.pop('name')  # Returns 'abc', d becomes {'age': 21}
```

````python
# Example: pop()
d = {'name': 'abc', 'age': 21, 'city': 'blr', 'phone': 1234}
print(f"Original dictionary: {d}")

removed_value = d.pop('name')
print(f"pop('name') returned: {removed_value}")
print(f"After pop('name'): {d}")

removed_value = d.pop('city')
print(f"\npop('city') returned: {removed_value}")
print(f"After pop('city'): {d}")

# Using pop with default value to avoid error
removed_value = d.pop('nonexistent', 'default value')
print(f"\npop('nonexistent', 'default') returned: {removed_value}")
print(f"Dictionary unchanged: {d}")

# Uncomment to see KeyError
# d.pop('nonexistent')  # This would raise KeyError
````

### 10. popitem()
Removes and returns the last inserted key-value pair as a tuple. Takes no arguments. Raises KeyError if dictionary is empty.

**Example:**
```python
d = {'name': 'abc', 'age': 21}
d.popitem()  # Returns ('age', 21), d becomes {'name': 'abc'}
```

````python
# Example: popitem()
d = {'name': 'abc', 'age': 21, 'city': 'blr', 'phone': 1234}
print(f"Original dictionary: {d}")

removed_item = d.popitem()
print(f"popitem() returned: {removed_item}")
print(f"After first popitem(): {d}")

removed_item = d.popitem()
print(f"\nSecond popitem() returned: {removed_item}")
print(f"After second popitem(): {d}")

removed_item = d.popitem()
print(f"\nThird popitem() returned: {removed_item}")
print(f"After third popitem(): {d}")

# Uncomment to see KeyError on empty dictionary
# d.clear()
# d.popitem()  # This would raise KeyError
````

## Dictionary Methods Summary

| Method | Description | Return Value | Modifies Original? |
|--------|-------------|--------------|-------------------|
| `copy()` | Shallow copy | New dictionary | No |
| `clear()` | Remove all items | None | Yes |
| `keys()` | Get all keys | dict_keys view | No |
| `values()` | Get all values | dict_values view | No |
| `items()` | Get all key-value pairs | dict_items view | No |
| `get(key, default)` | Get value for key | Value or default | No |
| `setdefault(key, default)` | Get value, set if missing | Value | Yes (if missing) |
| `update(other)` | Update with other dict | None | Yes |
| `pop(key, default)` | Remove and return value | Removed value | Yes |
| `popitem()` | Remove and return last item | (key, value) tuple | Yes |
