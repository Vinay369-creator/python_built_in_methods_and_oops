# List Built-in Methods in Python

This notebook covers all the important built-in methods available for lists in Python, with examples and explanations.

## Basic List Operations

### 1. count(val)
Returns the number of occurrences of a given value in the list. If the value is not present, it returns 0 (not an error).

**Example:**
```python
l = [1, 4, 55, 3, 1]
l.count(1)  # Output: 2
```

````python
# Example: count()
l = [1, 4, 55, 3, 1]
print(f"Original list: {l}")
print(f"count(1): {l.count(1)}")
print(f"count(55): {l.count(55)}")
print(f"count(100): {l.count(100)}")  # Returns 0, not an error
````

### 2. index(val, [start], [end])
Returns the first matched index position of the given value. If the value is not present, it raises a ValueError.

**Example:**
```python
l = [1, 4, 55, 3, 1]
l.index(1)  # Output: 0
```

````python
# Example: index()
l = [1, 4, 55, 3, 1]
print(f"Original list: {l}")
print(f"index(1): {l.index(1)}")
print(f"index(1, 1, 4): {l.index(1, 1, 4)}")

# Uncomment to see ValueError
# print(l.index(100))
````

### 3. clear()
Deletes all elements in the list. The list becomes empty but the memory is not erased.

**Example:**
```python
l = [1, 4, 55, 3, 1]
l.clear()  # Output: []
```

````python
# Example: clear()
l = [1, 4, 55, 3, 1]
print(f"Before clear: {l}")
l.clear()
print(f"After clear: {l}")
````

## Adding Elements to Lists

### 4. append(cdt/value)
Adds an element at the end of the list. It can take both collection data types and single values.

**Examples:**
```python
l = [1, 3]
l.append(10)   # [1, 3, 10]
l.append('hai') # [1, 3, 10, 'hai']
```

````python
# Example: append()
l = [1, 3]
print(f"Original list: {l}")

l.append(10)
print(f"After append(10): {l}")

l.append('hai')
print(f"After append('hai'): {l}")

l.append([4, 5])
print(f"After append([4, 5]): {l}")  # Appends entire list as one element
````

### 5. extend(cdt)
Adds elements at the end of the list. It takes only collection data types and extracts each element from that collection and adds them individually.

**Examples:**
```python
l = [1, 3]
l.extend([10])     # [1, 3, 10]
l.extend('hai')    # [1, 3, 'h', 'a', 'i']
```

````python
# Example: extend()
l = [1, 3]
print(f"Original list: {l}")

l.extend([10])
print(f"After extend([10]): {l}")

l.extend('hai')
print(f"After extend('hai'): {l}")

# Compare with append
l2 = [1, 3]
l2.append([10])
print(f"append([10]) vs extend([10]): {l2} vs {[1,3,10]}")
````

### 6. insert(index pos, cdt/value)
Adds an element at a specified index position.
- If index > list length, element is added at the end
- If index < 0, element is added at the beginning

**Examples:**
```python
l = [1, 3]
l.insert(1, 10)   # [1, 10, 3]
l.insert(3, 20)   # [1, 10, 3, 20]
```

````python
# Example: insert()
l = [1, 3]
print(f"Original list: {l}")

l.insert(1, 10)
print(f"After insert(1, 10): {l}")

l.insert(3, 20)  # Index 3 is beyond current length
print(f"After insert(3, 20): {l}")

l.insert(-1, 99)  # Negative index
print(f"After insert(-1, 99): {l}")
````

## Removing Elements from Lists

### 7. pop([index pos])
Deletes and returns an element. By default, it removes the last element. If index is out of range, it raises an IndexError.

**Examples:**
```python
l = [1, 3, 10, 20]
l.pop()    # Returns 20, list becomes [1, 3, 10]
l.pop(1)   # Returns 3, list becomes [1, 10]
```

````python
# Example: pop()
l = [1, 3, 10, 20]
print(f"Original list: {l}")

popped = l.pop()
print(f"pop() returned: {popped}")
print(f"After pop(): {l}")

popped = l.pop(1)
print(f"pop(1) returned: {popped}")
print(f"After pop(1): {l}")
````

### 8. remove(value)
Removes the first occurrence of the given value. If the value is not present, it raises a ValueError. It does not return the deleted element.

**Examples:**
```python
l = [1, 3, 10]
l.remove(1)   # l becomes [3, 10]
l.remove(33)  # ValueError
```

````python
# Example: remove()
l = [1, 3, 10, 1, 5]
print(f"Original list: {l}")

l.remove(1)
print(f"After remove(1): {l}")  # Removes first occurrence only

l.remove(10)
print(f"After remove(10): {l}")

# Uncomment to see ValueError
# l.remove(33)
````

## Sorting Lists

### 9. sort()
Sorts the elements in the list in-place. It doesn't return anything. By default, it sorts in ascending order.

**Examples:**
```python
l = [3, 5, 1]
l.sort()  # [1, 3, 5]

# Descending order
l.sort(reverse=True)  # [5, 3, 1]
```

````python
# Example: sort()
l = [3, 5, 1, 8, 2]
print(f"Original list: {l}")

l.sort()
print(f"After sort(): {l}")

l.sort(reverse=True)
print(f"After sort(reverse=True): {l}")

# Sorting strings
words = ['banana', 'apple', 'cherry']
words.sort()
print(f"Sorted words: {words}")
````

## Copying Lists (Important Concept)

### 10. Normal Copy Method (Assignment)
In normal copy (assignment), both the outer object and inner objects share the same memory. Changes made by the copied list directly affect the original list.

**Example:**
```python
l = [1, 2, 3, ['hai', 10], 44]
ncl = l  # Normal copy (assignment)
ncl.remove(3)  # Affects both lists
```

````python
# Example: Normal Copy (Assignment)
l = [1, 2, 3, ['hai', 10], 44]
print(f"Original list l: {l}")

# Normal copy (assignment)
ncl = l
print(f"Copied list ncl: {ncl}")

# Modify the copied list
ncl.remove(3)
print("\nAfter ncl.remove(3):")
print(f"Original list l: {l}")
print(f"Copied list ncl: {ncl}")

# Modify nested list through copied list
ncl[2][1] = 89  # Index 2 now points to ['hai', 10] after removal
print("\nAfter ncl[2][1] = 89:")
print(f"Original list l: {l}")
print(f"Copied list ncl: {ncl}")
````

### 11. Shallow Copy Method
In shallow copy, it copies values from the outer object but shares memory with inner objects. Changes in inner objects affect both lists. Use the `copy()` method.

**Example:**
```python
l = [1, 2, 3, ['hai', 10], 44]
scl = l.copy()
l[3][1] = 89  # Affects both lists for inner objects
```

````python
# Example: Shallow Copy
l = [1, 2, 3, ['hai', 10], 44]
print(f"Original list l: {l}")

# Shallow copy
scl = l.copy()
print(f"Shallow copy scl: {scl}")

# Modify inner object in original list
l[3][1] = 89
print("\nAfter l[3][1] = 89:")
print(f"Original list l: {l}")
print(f"Shallow copy scl: {scl}")  # Inner object changes affect both

# Modify outer object in original list
l.remove(3)
print("\nAfter l.remove(3):")
print(f"Original list l: {l}")
print(f"Shallow copy scl: {scl}")  # Outer object changes don't affect copy
````

### 12. Deep Copy Method
In deep copy, it copies values from both outer and inner objects. Changes in the original list do not affect the copied list. Requires the `copy` module.

**Example:**
```python
import copy
l = [1, 2, 3, ['hai', 10], 44]
dcl = copy.deepcopy(l)
l[3][1] = 89  # Only affects original list
```

````python
# Example: Deep Copy
import copy

l = [1, 2, 3, ['hai', 10], 44]
print(f"Original list l: {l}")

# Deep copy
dcl = copy.deepcopy(l)
print(f"Deep copy dcl: {dcl}")

# Modify inner object in original list
l[3][1] = 89
print("\nAfter l[3][1] = 89:")
print(f"Original list l: {l}")
print(f"Deep copy dcl: {dcl}")  # Inner object changes don't affect deep copy

# Modify outer object in original list
l.remove(3)
print("\nAfter l.remove(3):")
print(f"Original list l: {l}")
print(f"Deep copy dcl: {dcl}")  # Outer object changes don't affect deep copy
````

## Comparison of Copy Methods

| Copy Method | How to Use | Outer Object | Inner Objects | Independent? |
|-------------|------------|--------------|---------------|--------------|
| Normal Copy | `new = original` | Shared memory | Shared memory | No |
| Shallow Copy | `new = original.copy()` | New memory | Shared memory | Partially |
| Deep Copy | `new = copy.deepcopy(original)` | New memory | New memory | Yes |

````python
# Complete comparison example
import copy

original = [1, 2, [3, 4]]
print(f"Original: {original}")

# Normal copy
normal = original

# Shallow copy
shallow = original.copy()

# Deep copy
deep = copy.deepcopy(original)

# Modify inner object in original
original[2][0] = 99
original[0] = 100

print("\nAfter modifications:")
print(f"Original: {original}")
print(f"Normal copy: {normal}")    # Both outer and inner changes reflected
print(f"Shallow copy: {shallow}")   # Only inner changes reflected
print(f"Deep copy: {deep}")         # No changes reflected
````
