## Tuple Methods

### 1. count(val)

Returns the number of occurrences of a given value in the tuple. If the value is not present, it returns 0.

**Syntax:**
```python
tuple.count(value)
```

**Example:**
```python
t = (1, 4, 55, 3, 1)
t.count(1)  # Returns 2
```

````python
# Basic count() examples
t = (1, 4, 55, 3, 1, 1, 4, 2, 1)
print(f"Tuple: {t}")
print(f"Length of tuple: {len(t)}")
print("-" * 30)

print(f"count(1): {t.count(1)}")  # Count occurrences of 1
print(f"count(4): {t.count(4)}")  # Count occurrences of 4
print(f"count(55): {t.count(55)}")  # Count occurrences of 55
print(f"count(100): {t.count(100)}")  # Value not present - returns 0
````

````python
# count() with mixed data types
t = (1, 'hello', 3.14, True, 'hello', 1, False, 'hello')
print(f"Mixed tuple: {t}")

print(f"count('hello'): {t.count('hello')}")
print(f"count(1): {t.count(1)}")  # Note: True == 1 in Python
print(f"count(True): {t.count(True)}")  # True counts as 1
print(f"count(False): {t.count(False)}")  # False counts as 0
print(f"count(3.14): {t.count(3.14)}")
````

### 2. index(val, [start], [end])

Returns the first matched index position of the given value. If the value is not present, it raises a ValueError.

**Syntax:**
```python
tuple.index(value, start, end)
```
- `value`: The element to search for
- `start` (optional): Starting index for search
- `end` (optional): Ending index for search

**Example:**
```python
t = (1, 4, 55, 3, 1)
t.index(1)  # Returns 0
```

````python
# Basic index() examples
t = (10, 20, 30, 40, 50, 30, 60, 30)
print(f"Tuple: {t}")

# Find first occurrence
print(f"index(30): {t.index(30)}")  # First 30 is at index 2
print(f"index(10): {t.index(10)}")  # First 10 is at index 0
print(f"index(60): {t.index(60)}")  # First 60 is at index 6

# Using start parameter
print(f"index(30, 3): {t.index(30, 3)}")  # Search from index 3
print(f"index(30, 5): {t.index(30, 5)}")  # Search from index 5

# Using both start and end parameters
print(f"index(30, 3, 7): {t.index(30, 3, 7)}")  # Search between indices 3-7
````

````python
# index() with strings
t = ('apple', 'banana', 'cherry', 'banana', 'date', 'banana')
print(f"Fruit tuple: {t}")

print(f"index('banana'): {t.index('banana')}")
print(f"index('banana', 2): {t.index('banana', 2)}")
print(f"index('banana', 3, 5): {t.index('banana', 3, 5)}")
print(f"index('cherry'): {t.index('cherry')}")
````

## Summary Table

| Method | Description | Syntax | Return Value | Example |
|--------|-------------|--------|--------------|---------|
| `count()` | Count occurrences of a value | `tuple.count(value)` | Integer | `(1,2,1).count(1) → 2` |
| `index()` | Find first index of a value | `tuple.index(value, start, end)` | Integer | `(1,2,3).index(2) → 1` |

## Quick Reference Card

```python
# Tuple Methods Quick Reference

t = (1, 2, 3, 2, 4, 2, 5)

# count() - Count occurrences
count_of_2 = t.count(2)        # Returns 3
count_of_9 = t.count(9)        # Returns 0 (no error)

# index() - Find position
first_2 = t.index(2)           # Returns 1
second_2 = t.index(2, 2)       # Returns 3 (search from index 2)
third_2 = t.index(2, 4, 6)     # Returns 5 (search between 4-6)
# t.index(9)                   # ValueError: tuple.index(x): x not in tuple

# Other common operations
length = len(t)                 # Tuple length
exists = 3 in t                 # Membership test (True)
element = t[0]                   # Access element (indexing)
sliced = t[1:4]                  # Slicing (returns new tuple)
```

## Key Takeaways

1. **Tuples have only two built-in methods**: `count()` and `index()`
2. **count()** returns the number of occurrences (0 if not found)
3. **index()** returns the first index position (raises ValueError if not found)
4. **Tuples are immutable** - they cannot be modified after creation
5. **Use tuples for fixed data** that shouldn't change (coordinates, RGB values, function returns)
6. **index() accepts optional start/end parameters** to limit the search range
7. **count() is safe** - it returns 0 instead of raising an error for missing values
