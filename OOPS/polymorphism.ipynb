# POLYMORPHISM in Python

## One of the Five Major Pillars of OOPS

## What is Polymorphism?

**Polymorphism** is a process of methods behaving in different ways from one object to another object.

### Key Concept:
- The word "polymorphism" means **"many forms"**
- Same method name, different implementations
- Different behaviors for different objects

## Two Ways to Achieve Polymorphism

We can achieve polymorphism in two ways:

1. **Method Overloading** (Compile-time polymorphism)
2. **Method Overriding** (Runtime polymorphism)

---

## 1. METHOD OVERLOADING

**Definition:** It is a process of defining the same method multiple times with different arguments in a single class.

### Important Note:
> **In Python, method overloading is not directly supported** like in Java or C++.

### How to Achieve Method Overloading in Python:
We can achieve method overloading by using **default arguments**.

````python
# METHOD OVERLOADING using default arguments

class Overloading:
    def __init__(self, a, b=0, c=0):
        self.a = a
        self.b = b
        self.c = c
    
    def addValues(self):
        return self.a + self.b + self.c

print("=== METHOD OVERLOADING EXAMPLE ===")
print("(Achieved using default arguments)\n")

# Creating objects with different number of arguments
obj1 = Overloading(1)
print(f"obj1 = Overloading(1) -> addValues(): {obj1.addValues()}")  # 1

obj2 = Overloading(1, 4)
print(f"obj2 = Overloading(1, 4) -> addValues(): {obj2.addValues()}")  # 5

obj3 = Overloading(1, 4, 6)
print(f"obj3 = Overloading(1, 4, 6) -> addValues(): {obj3.addValues()}")  # 11
````

````python
# Another Method Overloading Example

class Calculator:
    def multiply(self, a, b=None, c=None):
        if b is None and c is None:
            return a  # Single argument
        elif c is None:
            return a * b  # Two arguments
        else:
            return a * b * c  # Three arguments

print("=== CALCULATOR OVERLOADING EXAMPLE ===")
calc = Calculator()

print(f"multiply(5): {calc.multiply(5)}")
print(f"multiply(5, 4): {calc.multiply(5, 4)}")
print(f"multiply(5, 4, 3): {calc.multiply(5, 4, 3)}")
````

## 2. METHOD OVERRIDING

**Definition:** It is a process of rewriting the same methods or properties of the parent class in the child class.

### Key Concept:
- Child class provides a specific implementation of a method that is already defined in parent class
- Method name and parameters remain the same
- Implementation differs

````python
# METHOD OVERRIDING - Basic Example

class Overriding:
    def displayOne(self):
        print('This is first class object method')

class Two(Overriding):
    def displayTwo(self):
        print('This is second class object method')

# Creating object of child class
obj = Two()
obj.displayOne()  # Inherited from parent
obj.displayTwo()  # Defined in child
````

````python
# METHOD OVERRIDING - Payment System Example

class Payment:
    amount = 200
    
    def payBill(self):
        print(f'Paying the bill amount {self.amount} through cash')

class PaymentCard(Payment):
    amount = 400
    
    def payBill(self):  # Method Overriding
        print(f'Paying the bill amount {self.amount} through Card')

class PaymentOnline(Payment):
    amount = 1500
    
    def payBill(self):  # Method Overriding
        print(f'Paying the bill amount {self.amount} through Online Banking')

print("=== METHOD OVERRIDING - PAYMENT EXAMPLE ===")

# Creating objects
cash = Payment()
card = PaymentCard()
online = PaymentOnline()

# Same method name, different behaviors
cash.payBill()
card.payBill()
online.payBill()

print("\n Same method 'payBill()' behaves differently for different objects")
````

````python
# METHOD OVERRIDING - Animal Sounds Example

class Animal:
    def sound(self):
        return "Some generic animal sound"

class Dog(Animal):
    def sound(self):  # Method Overriding
        return "Bark! Bark!"

class Cat(Animal):
    def sound(self):  # Method Overriding
        return "Meow! Meow!"

class Cow(Animal):
    def sound(self):  # Method Overriding
        return "Moo! Moo!"

print("=== METHOD OVERRIDING - ANIMAL SOUNDS ===")

animals = [Dog(), Cat(), Cow()]

for animal in animals:
    print(f"{animal.__class__.__name__} makes sound: {animal.sound()}")
````

---

## Disadvantages of Method Overriding

1. **Code redundancy will be more** - Duplicate code in multiple classes
2. **Memory consumption will be more** - Each class stores its own methods
3. **Reduces code efficiency** - More code to execute
4. **Application becomes slower** - Additional method lookups

### Example showing redundancy:

````python
# Problem: Code Redundancy in Method Overriding

class Logger:
    def log(self, message):
        timestamp = "2024-01-01 10:00:00"  # Redundant in every child
        return f"[{timestamp}] {message}"

class FileLogger(Logger):
    def log(self, message):
        timestamp = "2024-01-01 10:00:00"  # REDUNDANT - same code
        return f"FILE: [{timestamp}] {message}"

class DatabaseLogger(Logger):
    def log(self, message):
        timestamp = "2024-01-01 10:00:00"  # REDUNDANT - same code
        return f"DB: [{timestamp}] {message}"

print("=== DISADVANTAGE: CODE REDUNDANCY ===")
print("The timestamp formatting code is duplicated in every class!")

file_log = FileLogger()
db_log = DatabaseLogger()

print(file_log.log("User logged in"))
print(db_log.log("User logged in"))
````

## Solution: Chaining Methods

To reduce the problems caused by method overriding, Python provides **chaining methods**.

**Chaining Methods:** These methods are used to call another class's object method or constructor inside the current class constructor or method.

### Two Ways to Achieve Chaining:
1. **Super Method**
2. **Class Name Method**

### 1. SUPER METHOD

**Definition:** It is a process of calling the parent class constructor or method inside the child class constructor or method.

**Characteristics:**
- Inheritance is **mandatory** to use super method
- Used with the `super()` keyword
- No need to pass the mandatory argument (self)

**Syntax:**
```python
super().__init__(arguments)
super().method_name(arguments)
```

````python
# SUPER METHOD Example

class Notifier:
    def __init__(self, recipient):
        self.recipient = recipient

    def send(self, message):
        return f"Notifier: Preparing to send a message to {self.recipient}"
           
class EmailNotifier(Notifier):
    def send(self, message):
        # Chaining the parent method using super()
        parent_status = super().send(message)
        email_body = f"{parent_status}\nEmail: Sending email with message: '{message}'"
        print(email_body)
        return email_body

class SMSNotifier(Notifier):
    def send(self, message):
        # Not using parent method - different implementation
        print(f"SMS: Sending SMS to {self.recipient} with message: '{message}'")

print("=== SUPER METHOD EXAMPLE ===")

email_user = EmailNotifier("user@example.com")
email_user.send("Your password reset link is here.")

print()
sms_user = SMSNotifier("+1234567890")
sms_user.send("Your verification code is 1234.")
````

````python
# SUPER METHOD with Constructor Example

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        print(f"Vehicle constructor: {brand} {model}")
    
    def start(self):
        return f"{self.brand} {self.model} is starting"

class Car(Vehicle):
    def __init__(self, brand, model, doors):
        # Calling parent constructor using super()
        super().__init__(brand, model)
        self.doors = doors
        print(f"Car constructor: {doors} doors")
    
    def start(self):
        # Calling parent method using super()
        parent_start = super().start()
        return f"{parent_start} with {self.doors} doors"

print("=== SUPER METHOD WITH CONSTRUCTOR ===")
my_car = Car("Toyota", "Camry", 4)
print(my_car.start())
````

### 2. CLASS NAME METHOD

**Definition:** It is a process of calling another class's method or constructor by using the class name.

**Characteristics:**
- Inheritance is **NOT mandatory**
- Must pass the mandatory argument (self)

**Syntax:**
```python
ClassName.method_name(self, arguments)
```

````python
# CLASS NAME METHOD Example

class Logger:
    def log(self, level, message):
        return f"[{level}] {message}"

class FileHandler:
    def __init__(self, filename):
        self.filename = filename
    
    def write(self, content):
        print(f"Writing to {self.filename}: {content}")

class Application:
    def __init__(self, name):
        self.name = name
        self.logger = Logger()
        self.file_handler = FileHandler("app.log")
    
    def process(self, data):
        # Using class name method (without inheritance)
        log_message = self.logger.log("INFO", f"Processing data in {self.name}")
        
        # Using another class method
        self.file_handler.write(log_message)
        
        return f"Processed: {data}"

print("=== CLASS NAME METHOD EXAMPLE ===")
print("(Inheritance NOT required)\n")

app = Application("MyApp")
result = app.process("user data")
print(result)
````

## Comparison Table: Method Overloading vs Method Overriding

| Feature | Method Overloading | Method Overriding |
|---------|-------------------|-------------------|
| **Definition** | Same method name, different parameters in same class | Same method name, same parameters in parent-child classes |
| **Relationship** | Within same class | Between parent and child classes |
| **Inheritance** | Not required | Required |
| **Python Support** | Not directly (use default args) | Fully supported |
| **Purpose** | To perform similar operations with different inputs | To change parent class behavior in child |
| **Method Signature** | Different number/type of parameters | Same parameters |
| **Compile/Runtime** | Compile-time polymorphism | Runtime polymorphism |

## Comparison Table: Super Method vs Class Name Method

| Feature | Super Method | Class Name Method |
|---------|--------------|-------------------|
| **Inheritance Required** | Yes | No |
| **Syntax** | `super().method()` | `ClassName.method(self)` |
| **Pass self** | No (automatically passed) | Yes (must pass explicitly) |
| **Use Case** | Accessing parent class members | Accessing any class members |
| **Flexibility** | Limited to parent class | Can access any class |

## Key Takeaways

1. **Polymorphism** means "many forms" - same method, different behaviors

2. **Method Overloading** (same class, different parameters):
   - Not directly supported in Python
   - Achieved using **default arguments**

3. **Method Overriding** (parent-child classes):
   - Child redefines parent method
   - Provides specific implementation

4. **Disadvantages of Overriding**:
   - Code redundancy
   - More memory consumption
   - Reduced efficiency

5. **Chaining Methods** solve overriding problems:
   - **Super Method**: `super().method()` - requires inheritance
   - **Class Name Method**: `ClassName.method(self)` - no inheritance needed

6. **Polymorphism Benefits**:
   - Code reusability
   - Flexibility
   - Easy maintenance
   - Extensibility

````python
# Final Complete Example: Payment System with Chaining

class TransactionLogger:
    def log(self, transaction_type, amount):
        return f"[LOG] {transaction_type}: ${amount}"

class Payment:
    def __init__(self, amount):
        self.amount = amount
        self.logger = TransactionLogger()
    
    def process(self):
        return f"Processing payment of ${self.amount}"

class CreditCardPayment(Payment):
    def __init__(self, amount, card_number):
        super().__init__(amount)  # Super method for constructor
        self.card_number = card_number
    
    def process(self):
        # Super method for parent method
        parent_process = super().process()
        
        # Class name method for logger
        log_message = self.logger.log("CREDIT CARD", self.amount)
        
        return f"{parent_process}\n{log_message}\nCard: {self.card_number[-4:]}"

class PayPalPayment(Payment):
    def __init__(self, amount, email):
        super().__init__(amount)
        self.email = email
    
    def process(self):
        parent_process = super().process()
        log_message = self.logger.log("PAYPAL", self.amount)
        
        return f"{parent_process}\n{log_message}\nEmail: {self.email}"

print("=" * 50)
print("COMPLETE PAYMENT SYSTEM WITH CHAINING")
print("=" * 50)

cc_payment = CreditCardPayment(100, "1234-5678-9012-3456")
print(cc_payment.process())

print("\n" + "-" * 50 + "\n")

paypal_payment = PayPalPayment(75, "user@example.com")
print(paypal_payment.process())

print("\n Same process() method, different implementations (Polymorphism)")
print(" Super method used for parent constructor and method")
print(" Class name method used for logger (no inheritance)")
````
