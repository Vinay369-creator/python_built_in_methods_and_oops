# INHERITANCE in Python

## One of the Five Major Pillars of OOPS

## What is Inheritance?

**Inheritance** is a process of accessing another class into the current class.

### Key Concept:
- Represents **"IS-A" relationship**
- Child class acquires properties and methods of parent class
- Promotes code reusability

### Syntax:
```python
class Parent_Class_Name:
    pass

class Child_Class_Name(Parent_Class_Name):
    pass
```

## Types of Inheritance

Inheritance is classified into **5 types**:

1. **Single Level Inheritance**
2. **Multiple Inheritance**
3. **Multilevel Inheritance**
4. **Hierarchical Inheritance**
5. **Hybrid Inheritance**

---

## 1. SINGLE LEVEL INHERITANCE

**Definition:** It is a process of accessing one single parent class inside one single child class.

### Syntax:
```python
class A:
    pass

class B(A):
    pass
```

````python
# SINGLE LEVEL INHERITANCE Example

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start_engine(self):
        return f"The {self.brand} {self.model}'s engine has started."

class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        # Call the parent class constructor using super()
        super().__init__(brand, model)
        self.fuel_type = fuel_type

    def drive(self):
        return f"The {self.brand} {self.model} is now driving using {self.fuel_type}."

print("=== SINGLE LEVEL INHERITANCE ===")
my_car = Car("Toyota", "Corolla", "Gasoline")

print(my_car.start_engine())  # Inherited from Vehicle
print(my_car.drive())         # Defined in Car
````

---

## 2. MULTIPLE INHERITANCE

**Definition:** It is a process of acquiring multiple parent classes into one single child class.

### Syntax:
```python
class A:
    pass

class B:
    pass

class C(A, B):
    pass
```

````python
# MULTIPLE INHERITANCE Example

class Wheels:
    def __init__(self, co, si, typ):
        self.count = co
        self.size = si
        self.type = typ
    
    def w_display(self):
        print(f'wheel count: {self.count}')
        print(f'wheel size: {self.size}')
        print(f'wheel type: {self.type}')

class Steering:
    def __init__(self, top, modes):
        self.type = top
        self.modes = modes
    
    def s_display(self):
        print(f'steering type: {self.type}')
        print(f'steering modes: {self.modes}')

class Bmw(Wheels, Steering):
    def __init__(self, model, pri, co, si, typ, top, modes):
        self.model = model
        self.price = pri
        Wheels.__init__(self, co, si, typ)
        Steering.__init__(self, top, modes)

    def b_dis(self):
        print(f"Model: {self.model}")
        print(f"Price: ${self.price}")
        self.w_display()
        self.s_display()

print("=== MULTIPLE INHERITANCE ===")
b1 = Bmw('BMW M5', 7600000, 4, 18, 'flat-run', 'M-Sport', ['normal', 'sports'])
b1.b_dis()
````

## 3. MULTILEVEL INHERITANCE

**Definition:** It is a process of acquiring one child's properties into another child class (chain of inheritance).

### Syntax:
```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

````python
# MULTILEVEL INHERITANCE Example

class Person:
    def __init__(self, name, age, mob):
        self.name = name
        self.age = age
        self.mob = mob

    def p_details(self):
        print(f'Name: {self.name}')
        print(f'Age: {self.age}')
        print(f'Mobile No: {self.mob}')

class Student(Person):
    def __init__(self, name, age, mob, cls, fat, add):
        super().__init__(name, age, mob)
        self.sclass = cls
        self.father = fat
        self.address = add

    def p_details(self):
        super().p_details()
        print(f'Class: {self.sclass}')
        print(f'Father: {self.father}')
        print(f'Address: {self.address}')

class Hostel(Student):
    def __init__(self, name, age, mob, cls, fat, add, rno, hfee, mfee, pfee):
        super().__init__(name, age, mob, cls, fat, add)
        self.roomno = rno
        self.hostelfee = hfee
        self.messfee = mfee
        self.pendingfee = pfee

    def p_details(self):
        super().p_details()
        print(f'Room No: {self.roomno}')
        print(f'Hostel Fee: ${self.hostelfee}')
        print(f'Mess Fee: ${self.messfee}')
        print(f'Pending Fee: ${self.pendingfee}')

print("=== MULTILEVEL INHERITANCE ===")
p1 = Hostel('Luffy', 23, 1234, 4, 'Vinsmok', 'North', 203, 45000, 3000, 2000)
p1.p_details()
````

## 4. HIERARCHICAL INHERITANCE

**Definition:** It is a process of acquiring one single parent class into multiple child classes.

### Syntax:
```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass
```

````python
# HIERARCHICAL INHERITANCE Example

class Account:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Insufficient funds or invalid amount.")

class SavingsAccount(Account):
    def __init__(self, account_number, balance=0, interest_rate=0.01):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        interest = self.balance * self.interest_rate
        self.deposit(interest)
        print(f"Interest applied: ${interest:.2f}")

class CurrentAccount(Account):
    def __init__(self, account_number, balance=0, overdraft_limit=1000):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Withdrawal limit exceeded.")

print("=== HIERARCHICAL INHERITANCE ===")
savings = SavingsAccount("SAV123", 500, 0.02)
current = CurrentAccount("CUR456", 200)

print(f"Savings Account Balance: ${savings.balance}")
print(f"Current Account Balance: ${current.balance}")

print("\nSavings Account Operations:")
savings.deposit(100)
savings.apply_interest()

print("\nCurrent Account Operations:")
current.withdraw(300)
current.withdraw(1000)
````

## 5. HYBRID INHERITANCE

**Definition:** It is a process of using more than one type of inheritance together.

### Syntax:
```python
class A:
    pass

class B:
    pass

class C(A, B):   # Multiple Inheritance
    pass

class D(A):      # Single Inheritance
    pass
```

````python
# HYBRID INHERITANCE Example

class Engine:
    def __init__(self, engine_type, horsepower):
        self.engine_type = engine_type
        self.horsepower = horsepower
    
    def engine_info(self):
        return f"Engine: {self.engine_type}, {self.horsepower} HP"

class Transmission:
    def __init__(self, transmission_type, gears):
        self.transmission_type = transmission_type
        self.gears = gears
    
    def transmission_info(self):
        return f"Transmission: {self.transmission_type}, {self.gears} gears"

class Vehicle(Engine, Transmission):  # Multiple Inheritance
    def __init__(self, brand, model, engine_type, horsepower, transmission_type, gears):
        self.brand = brand
        self.model = model
        Engine.__init__(self, engine_type, horsepower)
        Transmission.__init__(self, transmission_type, gears)
    
    def vehicle_info(self):
        return f"Vehicle: {self.brand} {self.model}"

class Car(Vehicle):  # Multilevel Inheritance
    def __init__(self, brand, model, engine_type, horsepower, transmission_type, gears, doors):
        super().__init__(brand, model, engine_type, horsepower, transmission_type, gears)
        self.doors = doors
    
    def car_info(self):
        info = f"{self.vehicle_info()}\n"
        info += f"{self.engine_info()}\n"
        info += f"{self.transmission_info()}\n"
        info += f"Doors: {self.doors}"
        return info

class Bike(Vehicle):  # Hierarchical Inheritance
    def __init__(self, brand, model, engine_type, horsepower, transmission_type, gears, has_gear):
        super().__init__(brand, model, engine_type, horsepower, transmission_type, gears)
        self.has_gear = has_gear
    
    def bike_info(self):
        info = f"{self.vehicle_info()}\n"
        info += f"{self.engine_info()}\n"
        info += f"{self.transmission_info()}\n"
        info += f"Has Gear: {'Yes' if self.has_gear else 'No'}"
        return info

print("=== HYBRID INHERITANCE ===")
print("(Combines Multiple, Multilevel, and Hierarchical Inheritance)\n")

my_car = Car("Toyota", "Camry", "V6", 300, "Automatic", 8, 4)
my_bike = Bike("Honda", "CBR", "Inline-4", 150, "Manual", 6, True)

print(my_car.car_info())
print("\n" + "-" * 40 + "\n")
print(my_bike.bike_info())
````

## Inheritance Types Summary

| # | Inheritance Type | Description | Syntax | Example |
|---|-----------------|-------------|--------|---------|
| 1 | **Single Level** | One parent, one child | `class B(A):` | Vehicle → Car |
| 2 | **Multiple** | Multiple parents, one child | `class C(A, B):` | Wheels + Steering → BMW |
| 3 | **Multilevel** | Chain of inheritance | `class C(B):` where `B(A)` | Person → Student → Hostel |
| 4 | **Hierarchical** | One parent, multiple children | `class B(A):`, `class C(A):` | Account → Savings, Current |
| 5 | **Hybrid** | Combination of multiple types | Mix of above | Vehicle → Car & Bike with Engine + Transmission |

## Advantages of Inheritance

1. **Code Reusability**: Write once, use multiple times
2. **Method Overriding**: Runtime polymorphism
3. **Extensibility**: Easy to extend existing code
4. **Maintainability**: Changes in parent reflect in children
5. **DRY Principle**: Don't Repeat Yourself
