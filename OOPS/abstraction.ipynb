# ABSTRACTION in Python

## One of the Five Major Pillars of OOPS

## What is Abstraction?

**Abstraction** is a process of hiding the internal functionality from the user.

### Key Concept:
- User can only know **WHAT** is happening
- User doesn't know **HOW** it is happening
- Showing only essential features, hiding implementation details

### Real-world Example:
- When you drive a car, you know what the accelerator, brake, and steering do
- You don't need to know how the engine works internally

## Achieving Abstraction in Python

In Python, we cannot achieve abstraction directly. We have to import the **abc module** (Abstract Base Class).

### Import Statement:
```python
from abc import ABC, abstractmethod
```

The abc module contains:
- **ABC class** - Abstract Base Class
- **abstractmethod decorator** - Used to declare abstract methods

## What is Abstract Class?

**Abstract Class** is a prototype class which specifies to the child class what implementations should be written in that class.

### Characteristics:
- It is a class with all basic implementations of abstract methods
- Cannot be instantiated (cannot create objects of abstract class)
- Serves as a blueprint for other classes
- Child classes must implement all abstract methods

## What is Abstract Method?

**Abstract Method** is a method without any implementation which consists of the `pass` keyword.

### Syntax:
```python
@abstractmethod
def method_name(self):
    pass
```

## Types of Abstract Methods

| # | Type | Decorator | Description |
|---|------|-----------|-------------|
| 1 | **Mandatory Abstract Method** | `@abstractmethod` | Implementation MUST be present in child class |
| 2 | **Non-Mandatory Abstract Method** | No decorator | Implementation may or may not be present in child class |

---

## 1. MANDATORY ABSTRACT METHOD

**Definition:** These methods are declared using the `@abstractmethod` decorator before the method declaration. They specify that implementations MUST be present in the child class.

### Characteristics:
- Decorated with `@abstractmethod`
- Must be implemented in all child classes
- If not implemented, child class also becomes abstract
- Cannot create object of class with abstract methods

````python
# MANDATORY ABSTRACT METHOD Example
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    
    @abstractmethod  # Mandatory Abstract Method
    def process_payment(self, amount):
        pass

    @abstractmethod  # Mandatory Abstract Method
    def check_balance(self):
        pass

class CreditCardPayment(PaymentGateway):
    def __init__(self, card_number):
        self.card_number = card_number
        self.balance = 1000

    # Implementing mandatory abstract methods
    def process_payment(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Credit Card payment of ${amount} processed successfully for card {self.card_number}!")
        else:
            print("Insufficient funds for credit card transaction.")

    def check_balance(self):
        print(f"Credit Card Balance: ${self.balance}")

print("=== MANDATORY ABSTRACT METHODS ===")
payment1 = CreditCardPayment(1234)
payment1.process_payment(500)
payment1.check_balance()
````

---

## 2. NON-MANDATORY ABSTRACT METHOD

**Definition:** These methods are declared directly using the `def` keyword. They do not specify whether implementations should be present in the child class.

### Characteristics:
- No decorator needed
- Can have default implementation in abstract class
- Child class may or may not override them
- Optional to implement

````python
# NON-MANDATORY ABSTRACT METHOD Example
from abc import ABC, abstractmethod

class Database(ABC):
    
    @abstractmethod
    def connect(self):  # Mandatory
        pass
    
    @abstractmethod
    def disconnect(self):  # Mandatory
        pass
    
    # Non-mandatory method (concrete method)
    def log_connection(self):
        print("Connection logged at:", __import__('datetime').datetime.now())

class MySQLDatabase(Database):
    def connect(self):
        print("Connecting to MySQL database...")
    
    def disconnect(self):
        print("Disconnecting from MySQL database...")
    
    # Optional: May or may not override log_connection
    def log_connection(self):
        print("MySQL: Connection established")

print("=== NON-MANDATORY METHODS ===")
mysql = MySQLDatabase()
mysql.connect()
mysql.log_connection()  # Overridden version
mysql.disconnect()
````

## Complete Payment Gateway Example

````python
# COMPLETE PAYMENT GATEWAY EXAMPLE
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    """Abstract class for payment gateways"""
    
    def __init__(self, gateway_name):
        self.gateway_name = gateway_name
        self.transaction_id = 1000
    
    # Mandatory abstract methods
    @abstractmethod
    def process_payment(self, amount):
        pass

    @abstractmethod
    def check_balance(self):
        pass
    
    # Non-mandatory methods (concrete methods)
    def generate_transaction_id(self):
        self.transaction_id += 1
        return f"TXN{self.transaction_id}"
    
    def log_transaction(self, amount, status):
        txn_id = self.generate_transaction_id()
        print(f"[{self.gateway_name}] Transaction {txn_id}: ${amount} - {status}")

class CreditCardPayment(PaymentGateway):
    def __init__(self, card_number, cvv):
        super().__init__("Credit Card")
        self.card_number = card_number
        self.cvv = cvv
        self.balance = 5000

    def process_payment(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Credit Card payment of ${amount} processed successfully!")
            self.log_transaction(amount, "SUCCESS")
            return True
        else:
            print(f"Insufficient funds for credit card transaction.")
            self.log_transaction(amount, "FAILED - Insufficient Funds")
            return False

    def check_balance(self):
        print(f"Credit Card Balance: ${self.balance}")
        return self.balance

credit = CreditCardPayment("1234-5678-9012-3456", "123")
print("\n--- PROCESSING PAYMENTS ---")
credit.process_payment(100)
print("\n--- CHECKING BALANCES ---")
credit.check_balance()
````

## Comparison: Abstract Class vs Concrete Class

| Feature | Abstract Class | Concrete Class |
|---------|---------------|----------------|
| **Instantiation** | Cannot create objects | Can create objects |
| **Abstract Methods** | Can have abstract methods | Cannot have abstract methods |
| **Purpose** | Blueprint for other classes | Complete implementation |
| **Inheritance** | Must be inherited | May or may not be inherited |
| **Method Implementation** | May have both abstract & concrete methods | All methods are concrete |

## Advantages of Abstraction

1. **Hides Complexity** - Users only see what's necessary
2. **Code Reusability** - Common code in abstract class
3. **Security** - Internal implementation hidden
4. **Flexibility** - Easy to add new implementations
5. **Maintainability** - Changes in implementation don't affect users
6. **Standardization** - Ensures all child classes follow same interface

## Key Takeaways

1. **Abstraction** = Hiding internal functionality, showing only what's necessary

2. **Import Required**: `from abc import ABC, abstractmethod`

3. **Abstract Class**:
   - Cannot be instantiated
   - Serves as blueprint
   - Contains abstract methods

4. **Abstract Method**:
   - Method with `pass` (no implementation)
   - Decorated with `@abstractmethod`
   - Must be implemented in child classes

5. **Two Types of Methods**:
   - **Mandatory** (`@abstractmethod`) - Must implement
   - **Non-mandatory** (regular methods) - Optional to override

6. **Benefits**:
   - User knows WHAT, not HOW
   - Consistent interface
   - Easy to extend
