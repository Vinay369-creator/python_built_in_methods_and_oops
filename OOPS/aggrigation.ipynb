# AGGREGATION in Python

## One of the Five Major Pillars of OOPS

## What is Aggregation?

**Aggregation** is a process of accessing another class's object into the current class object.

### Key Concept:
- Aggregation is used when there is a **HAS-A relationship**

### Real-world Examples:
1. **Driver HAS-A Car**
2. **Team HAS-A Players**

## Two Scenarios for Using Another Class Object

There are two possible scenarios to use another class object into the current class object:

1. **Scenario 1**: The another class object is already created in main space and that object will be passed as an argument to the current class object creation.

2. **Scenario 2**: Creating the another class object inside the current class object constructor.

---

## Scenario 1: Object Created in Main Space

```python
# SCENARIO 1: Object created in main space and passed as argument

class Add:
    def __init__(self, val):
        self.val = val
    
    def disadd(self):
        print(self.val)

# Creating Add object in main space
addobj = Add('Bengaluru')

class Student:
    def __init__(self, name, age, add):
        self.name = name
        self.age = age
        self.add = add  # It holds the entire addobj

    def disstd(self):
        print(self.name)
        print(self.age)
        self.add.disadd()  # Calling the addobj method to access its properties

# Creating Student object with the addobj from main space
stdobj = Student('tony', 25, addobj)

print("=== SCENARIO 1: Object Created in Main Space ===")
stdobj.disstd()
```

### Drawbacks of Scenario 1

1. **Static Values**: The value will be static for the current class multiple objects.
2. **Unnecessary Memory Storage**: Unnecessary storage in memory if the another class object's properties are not used.

```python
# Demonstrating Drawback 1: Static Values

print("=== DRAWBACK 1: Static Values ===")
print("All objects share the SAME address value:\n")

# Single address object
single_add = Add('Bengaluru')

# Multiple students sharing the SAME address
s1 = Student('Student1', 20, single_add)
s2 = Student('Student2', 21, single_add)
s3 = Student('Student3', 22, single_add)

s1.disstd()
print()
s2.disstd()
print()
s3.disstd()

print("\n➡️ PROBLEM: All students show the same city!")
```

```python
# Demonstrating Drawback 2: Memory Waste

print("=== DRAWBACK 2: Memory Waste ===")

# Creating an object that will never be used
unused_obj = Add('This will waste memory')
print(f"unused_obj created with value: {unused_obj.val}")
print("But this object is never used by any Student!\n")

class AnotherClass:
    def __init__(self, name):
        self.name = name
        # Note: unused_obj is NOT used here
    
    def display(self):
        print(f"Name: {self.name}")

obj = AnotherClass("Test")
obj.display()

print("\n➡️ PROBLEM: unused_obj still occupies memory unnecessarily!")
```

## Scenario 2: Object Created Inside Constructor

```python
# SCENARIO 2: Creating object inside the constructor

class Add:
    def __init__(self, val):
        self.val = val

    def disadd(self):
        print(self.val)

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
        # Taking input for dynamic value
        val = input(f"Enter city for {name}: ")
        
        # Creating Add object inside constructor
        addobj = Add(val)
        self.add = addobj  # Aggregation

    def disstd(self):
        print(self.name)
        print(self.age)
        self.add.disadd()  # Calling the addobj method to access its properties

# Each student gets their OWN address object with dynamic values
stdobj1 = Student('tony', 25)
stdobj2 = Student('steve', 30)

print("\n" + "-"*30)
print("STUDENT DETAILS:")
print("-"*30)
stdobj1.disstd()
print()
stdobj2.disstd()
```

### Advantages of Scenario 2

1. **Dynamic Values**: Creates dynamic values during every object creation.
2. **No Memory Waste**: There is no waste of memory.

```python
# Demonstrating Advantage 1: Dynamic Values

class Address:
    def __init__(self, city):
        self.city = city
    
    def display(self):
        return self.city

class Employee:
    def __init__(self, name, role):
        self.name = name
        
        # Dynamic values based on role
        if role == "Manager":
            self.address = Address("Corporate Office")
        elif role == "Engineer":
            self.address = Address("Tech Park")
        else:
            self.address = Address("Branch Office")
    
    def display(self):
        print(f"{self.name} works at: {self.address.display()}")

print("=== ADVANTAGE 1: Dynamic Values ===")
e1 = Employee("John", "Manager")
e2 = Employee("Alice", "Engineer")
e3 = Employee("Bob", "Staff")

e1.display()
e2.display()
e3.display()
```

---

## Summary Table

| Feature | Scenario 1 (Object in Main) | Scenario 2 (Object in Constructor) |
|---------|----------------------------|-----------------------------------|
| **Object Creation** | In main space | Inside constructor |
| **Values** | Static - same for all objects | Dynamic - different for each object |
| **Memory Usage** | Potential waste | Optimal - created only when needed |
| **Flexibility** | Less flexible | More flexible |
| **Independence** | Objects share same instance | Each object has its own instance |

## Key Takeaways

1. **Aggregation** represents a **HAS-A relationship**

2. **Scenario 1** (object in main space):
   - Object created once in main space
   - Passed as argument to multiple objects
   - ❌ Static values for all objects
   - ❌ Potential memory waste

3. **Scenario 2** (object in constructor):
   - Object created inside constructor
   - Each object gets its own instance
   - ✅ Dynamic values per object
   - ✅ No memory waste

4. **Choose Scenario 2** when you need:
   - Different values for different objects
   - Memory efficiency
   - Object independence
