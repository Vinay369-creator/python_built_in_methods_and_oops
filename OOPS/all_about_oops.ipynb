# Object-Oriented Programming (OOPS) in Python

This notebook covers the fundamental concepts of Object-Oriented Programming in Python, including classes, objects, methods, and decorators.

## Introduction to OOPS

**Python supports multi-paradigm programming**, which means we can write code using functions (procedural programming) and by creating classes and objects (object-oriented programming).

**OOPS** stands for **Object-Oriented Programming Structure**. It is one of the popular approaches to solve real-world problems by creating objects and using them.

### Key Concepts:
- In Python, **everything is treated as an object**
- OOPS follows the **DRY concept** (Don't Repeat Yourself)

### Three Simple Steps in OOPS:
1. **Create a class** - Blueprint for objects
2. **Create objects** using the class
3. **Use the objects** to perform operations

## Class

A **class** is a blueprint or template for creating objects. It defines the properties (attributes) and behaviors (methods) that its objects will have.

**Syntax:**
```python
class ClassName:
    # properties (attributes)
    # states (methods)
```

When a class is created, a dictionary is created under the class name where:
- Variables/properties are represented with **key names**
- Values/functionality are represented with **value pairs**

````python
# Creating a simple class
class Student:
    # Class-level dictionary is created automatically
    pass

# Check the class dictionary
print(f"Student class dictionary: {dir(Student)}")
print(f"Student class name: {Student.__name__}")
````

## Objects

An **object** is an entity that contains states and behavior. The entity can be living or non-living.
- **States** represent the properties/attributes of the object
- **Behavior** represents the functionality/methods of the object

**Creating an object:**
```python
object_name = ClassName(args)
```

When an object is created, a dictionary is created under the object name. All variables/properties and values from the class are inherited by the object.

````python
# Creating a class and objects
class Car:
    pass

# Creating objects (instances) of the Car class
car1 = Car()
car2 = Car()

print(f"car1 object: {car1}")
print(f"car2 object: {car2}")
print(f"Type of car1: {type(car1)}")
````

## Dot Operator/Notation

The dot (`.`) is used to access properties and methods from both classes and objects.

````python
# Demonstrating dot operator
class Person:
    species = "Human"  # Class property
    
    def greet(self):    # Class method
        return "Hello!"

# Accessing class properties and methods using dot operator
print(f"Access via class - species: {Person.species}")
print(f"Access via class - greet: {Person.greet}")

# Creating and accessing via object
p1 = Person()
print(f"Access via object - species: {p1.species}")
print(f"Access via object - greet(): {p1.greet()}")
````

## Types of Properties

### 1. Generic Properties (Class Attributes)

These are variables created inside the class space, before any object methods.

**Characteristics:**
- Same for all objects
- Can be accessed by both class and objects
- If changed, it affects both class and objects
- **Cannot be changed using objects**

**Example:**
```python
class Generic:
    var1 = 'hello'
    var2 = 'bye'

obj1 = Generic()
print(Generic.var1)  # hello
print(obj1.var1)     # hello
```

````python
# Example: Generic Properties
class Generic:
    var1 = 'hello'
    var2 = 'bye'

print("=== Generic Properties ===")
obj1 = Generic()
obj2 = Generic()

print(f"Class access - Generic.var1: {Generic.var1}")
print(f"Object1 access - obj1.var1: {obj1.var1}")
print(f"Object2 access - obj2.var1: {obj2.var1}")

# Changing generic property via class
Generic.var1 = 'modified hello'
print("\nAfter changing Generic.var1:")
print(f"Class access - Generic.var1: {Generic.var1}")
print(f"Object1 access - obj1.var1: {obj1.var1}")
print(f"Object2 access - obj2.var1: {obj2.var1}")

# Trying to change via object (this creates an instance attribute instead)
obj1.var1 = 'object specific'
print("\nAfter obj1.var1 = 'object specific':")
print(f"Class access - Generic.var1: {Generic.var1}")
print(f"Object1 access - obj1.var1: {obj1.var1}")
print(f"Object2 access - obj2.var1: {obj2.var1}")
````

### 2. Specific Properties (Instance Attributes)

These are properties that purely belong to individual objects. They are declared inside a special method called `__init__`.

#### The `__init__` Method:
- Belongs to the class
- Starts executing whenever an object is created
- The interpreter implicitly calls this method
- Takes one mandatory argument called `self`
- `self` is responsible for pointing to the current instance

**Syntax:**
```python
class Example:
    def __init__(self, data):
        self.data = data
```

````python
# Example: Specific Properties with __init__
class Example1:
    name = 'this is generic'  # Generic property
    
    def __init__(self, data):
        self.data = data      # Specific property
        print(f"__init__ called for object with data: {data}")

# Creating objects - __init__ is automatically called
obj1 = Example1('this is specific properties 1')
obj2 = Example1('this is specific properties 2')

print("\n=== Accessing Properties ===")
print(f"obj1.data: {obj1.data}")           # Specific to obj1
print(f"obj2.data: {obj2.data}")           # Specific to obj2
print(f"Example1.name: {Example1.name}")   # Generic property
````

## Types of Methods

In Python classes, there are three types of methods:
1. **Object Methods** (Instance Methods)
2. **Class Methods**
3. **Static Methods**

### 1. Object Methods (Instance Methods)

These are methods created inside the class with the `def` keyword. They are directly accessible using the object.

**Syntax:**
- Via object: `object_name.method_name(args)`
- Via class: `class_name.method_name(object_instance, args)`

No special decorators are required to create object methods.

````python
# Example: Object Methods
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # Object method (instance method)
    def display_info(self):
        return f"Name: {self.name}, Age: {self.age}"
    
    def is_adult(self):
        return self.age >= 18

# Creating objects
s1 = Student("Alice", 20)
s2 = Student("Bob", 16)

# Accessing object methods via objects
print("=== Access via objects ===")
print(f"s1.display_info(): {s1.display_info()}")
print(f"s2.display_info(): {s2.display_info()}")
print(f"s1.is_adult(): {s1.is_adult()}")
print(f"s2.is_adult(): {s2.is_adult()}")

# Accessing object methods via class (passing object as argument)
print("\n=== Access via class ===")
print(f"Student.display_info(s1): {Student.display_info(s1)}")
````

### 2. Class Methods

Class methods are used to modify generic properties (class attributes). They can be accessed by both class and objects.

**Characteristics:**
- Created using the `@classmethod` decorator
- Take one mandatory argument called `cls` (referring to the class)
- Can modify class-level attributes

**Syntax:**
```python
@classmethod
def method_name(cls, args):
    # method body
```

````python
# Example: Class Methods
class Example2:
    name = 'generic'  # Generic property
    
    def __init__(self, val):
        self.val = val  # Specific property
    
    @classmethod
    def changeGenericProp(cls, val):
        cls.name = val
        print(f"Generic property changed to: {val}")
    
    @classmethod
    def display_class_info(cls):
        return f"Class name: {cls.__name__}, Generic property: {cls.name}"

print("=== Before using class method ===")
print(f"Example2.name: {Example2.name}")

obj1 = Example2('specific value')
print(f"obj1.val: {obj1.val}")
print(f"obj1.name: {obj1.name}")

print("\n=== After using class method ===")
Example2.changeGenericProp('this is class method example')
print(f"Example2.name: {Example2.name}")
print(f"obj1.name: {obj1.name}")  # Object reflects the change

# Class method accessed via object
print(f"\nobj1.display_class_info(): {obj1.display_class_info()}")
````

### 3. Static Methods

Static methods are utility methods that don't depend on class or instance state. They neither take `self` nor `cls` as mandatory arguments.

**Characteristics:**
- Created using the `@staticmethod` decorator
- Don't take any mandatory arguments
- Can be accessed by both class and objects
- Used to deliver required values during object creation or class modifications

**Syntax:**
```python
@staticmethod
def method_name(args):
    # method body
```

````python
# Example: Static Methods
class Example3:
    name = 'this is generic prop'
    
    def __init__(self, val):
        self.val = val
    
    @classmethod
    def changeGenericProp(cls):
        cls.name = cls.staticVal()
    
    def changeSpecific(self):
        self.val = self.staticVal()  # Using static method
    
    @staticmethod
    def staticVal():
        # This method doesn't depend on class or instance
        return "Value from static method"
    
    @staticmethod
    def add_numbers(a, b):
        return a + b

# Creating object
obj1 = Example3('this is specific')

print("=== Before calling methods ===")
print(f"obj1.val: {obj1.val}")
print(f"Example3.name: {Example3.name}")

# Using static methods
print("\n=== Static method usage ===")
print(f"Example3.add_numbers(10, 20): {Example3.add_numbers(10, 20)}")
print(f"obj1.add_numbers(5, 7): {obj1.add_numbers(5, 7)}")

# Using class method that uses static method
Example3.changeGenericProp()
print("\n=== After calling changeGenericProp ===")
print(f"Example3.name: {Example3.name}")

# Using instance method that uses static method
obj1.changeSpecific()
print(f"obj1.val: {obj1.val}")
````

### Interactive Static Method Example

````python
# Interactive example with user input
class UserManager:
    users_count = 0
    
    def __init__(self, username):
        self.username = username
        UserManager.users_count += 1
    
    @classmethod
    def create_from_input(cls):
        username = cls.get_user_input("Enter username: ")
        return cls(username)
    
    @staticmethod
    def get_user_input(prompt):
        # This static method handles user input
        return input(prompt)
    
    @staticmethod
    def validate_age(age):
        return 0 <= age <= 150

print("=== Creating users ===")
# Using static method for validation
age = 25
print(f"Is age {age} valid? {UserManager.validate_age(age)}")

# Uncomment to test interactive creation
# user1 = UserManager.create_from_input()
# print(f"Created user: {user1.username}")
# print(f"Total users: {UserManager.users_count}")
````

## Method Comparison Table

| Feature | Object Method | Class Method | Static Method |
|---------|---------------|--------------|---------------|
| Decorator | None | `@classmethod` | `@staticmethod` |
| Mandatory argument | `self` (instance) | `cls` (class) | None |
| Can access instance attributes | Yes | No | No |
| Can access class attributes | Yes | Yes | Yes (via class) |
| Can modify instance attributes | Yes | No | No |
| Can modify class attributes | Yes (via class) | Yes | No |
| Called via instance | Yes | Yes | Yes |
| Called via class | Yes (needs instance) | Yes | Yes |

## Decorators

**Decorators** are used to access, modify, and restrict the functionality of functions or classes.

### Key Terminology

1. **Decorator Function**: A function with one argument that has an inner function and returns the inner function's address
2. **Decorated Class/Function**: The class/function to which we apply the decoration
3. **Decorating**: The process of applying decoration using the decorator function

Decorators are functions used to swap function addresses.

### Types of Decorators

#### 1. Built-in Decorators
These are decorators available in Python:
- `@classmethod` - Converts a method to class method
- `@staticmethod` - Converts a method to static method
- `@property` - Used to define getters/setters
- `@abstractmethod` - For abstract base classes

#### 2. User-defined Decorators
These are decorators created by users based on their requirements.

### Creating User-defined Decorators

````python
# Basic decorator structure
def outer(arg):  # 'arg' holds the function being decorated
    print('This is outer function')
    
    def inner():
        print('This is the inner function')
        arg(10)  # Calling the decorated function with argument
    
    return inner  # Returning inner function address

@outer  # This is equivalent to: second = outer(second)
def second(a):
    print(f'This is now the decorated function with value: {a}')

# When we call second(), we're actually calling inner()
print("=== Calling decorated function ===")
second()
````
