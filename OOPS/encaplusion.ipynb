# ENCAPSULATION in Python

## One of the Five Major Pillars of OOPS

## What is Encapsulation?

**Encapsulation** is a process of binding the object's properties and methods into one single roof (class).

### Key Concept:
- It is used to **protect the data** with the help of **specifiers/modifiers**
- Bundling data and methods that operate on that data within a single unit
- Restricting direct access to some of an object's components

## Specifiers/Modifiers

These specifiers provide the **scope visibility** to the object variables.

### Three Types of Specifiers:

| # | Specifier | Symbol | Scope |
|---|-----------|--------|-------|
| 1 | **PUBLIC** | No underscore | Anywhere in Python (across packages) |
| 2 | **PROTECTED** | Single underscore `_` | Within the same package only |
| 3 | **PRIVATE** | Double underscore `__` | Within the same class only |

---

## 1. PUBLIC SPECIFIERS

**Definition:** These specifiers allow object properties to be used anywhere in Python - from one package to another package.

### Syntax:
```python
# For variables
variable_name = value

# For methods
def method_name(self):
    # method body
```

### Access:
We can access these variables directly using the object reference.
```python
# For variables
objectname.variable_name

# For methods
objectname.method_name()
```

````python
# PUBLIC SPECIFIERS Example

class Public:
    def __init__(self, data):
        self.data = data  # Public variable
    
    def display(self):    # Public method
        print(self.data)

print("=== PUBLIC SPECIFIERS ===")
obj1 = Public(10)

# Accessing public variable directly
print(f"Accessing public variable obj1.data: {obj1.data}")

# Accessing public method directly
print("Calling public method:", end=" ")
obj1.display()
````

---

## 2. PROTECTED SPECIFIERS

**Definition:** These specifiers allow object properties to be used only within that package. They cannot be used in other packages.

**Representation:** Single underscore (`_`)

### Syntax:
```python
# For variables
_variable_name = value

# For methods
def _method_name(self):
    # method body
```

### Access:
We can access these variables by using the underscore before the variable/method name with the object reference.
```python
# For variables
objectname._variable_name

# For methods
objectname._method_name()
```

````python
# PROTECTED SPECIFIERS Example

class Protect:
    def __init__(self, data, val):
        self._data = data  # Protected variable
        self._val = val    # Protected variable
    
    def _display(self):    # Protected method
        print(self._val)

print("=== PROTECTED SPECIFIERS ===")
obj1 = Protect(10, 'proct')

# Accessing protected variable (with underscore)
print(f"Accessing protected variable obj1._data: {obj1._data}")

# Accessing protected method (with underscore)
print("Calling protected method:", end=" ")
obj1._display()
````

````python
# Protected Access in Different Package (Conceptual)

print("=== PROTECTED SCOPE ===")
print("Protected members can be accessed:")
print("✅ Within the same class")
print("✅ Within the same package/module")
print("❌ Cannot be accessed from other packages")
````

---

## 3. PRIVATE SPECIFIERS

**Definition:** These specifiers allow object properties to be used only within that class space. They cannot be used outside the class.

**Representation:** Double underscore (`__`)

### Syntax:
```python
# For variables
__variable_name = value

# For methods
def __method_name(self):
    # method body
```

### Access:
We **cannot** access these variables directly using `__` with object reference.

We have to use: **object_name._classname__variable_name**

```python
# For variables
objectname._classname__variable_name

# For methods
objectname._classname__method_name()
```

````python
# PRIVATE SPECIFIERS Example

class Private:
    def __init__(self, data, val):
        self.__data = data  # Private variable
        self.__val = val    # Private variable
    
    def __display(self):    # Private method
        print(self.__val)

print("=== PRIVATE SPECIFIERS ===")
obj1 = Private(10, 'proct')

# Trying to access directly (will cause error)
try:
    print(obj1.__data)
except AttributeError as e:
    print(f"Direct access obj1.__data fails: {e}")

# Correct way to access private members
print(f"\nAccessing private variable obj1._Private__data: {obj1._Private__data}")

print("Calling private method:", end=" ")
obj1._Private__display()
````

### Important Note on Private Specifiers

> **Note:** By using these specifiers, we can achieve data protection, but **not 100%** protection. Private members can still be accessed using name mangling.

````python
# Private Members - Return and Use in Main Space

class Private:
    def __init__(self, data):
        self.__data = data
    
    def __display(self):
        return self.__data
    
    def get_data(self):  # Public method to access private data
        return self.__data

print("=== RETURNING PRIVATE VALUES ===")
obj1 = Private(10)

# Method 1: Using name mangling (direct access)
print(f"Using name mangling: obj1._Private__data = {obj1._Private__data}")

# Method 2: Using public method to return private value
val = obj1.get_data()
print(f"Using public getter method: {val}")

# Method 3: Accessing private method via name mangling
retval = obj1._Private__display()
print(f"Private method returned: {retval}")

print("\n➡️ Note: Private values can be returned and used in main space")
````

## Complete Example: Bank Account with Encapsulation

````python
# Bank Account Example demonstrating all specifiers

class BankAccount:
    def __init__(self, name, bal=0):
        self.name = name           # PUBLIC - can be accessed anywhere
        self._branch = "Main"      # PROTECTED - within package
        self.__balance = bal       # PRIVATE - within class only

    def deposit(self, amount):     # PUBLIC method
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):    # PUBLIC method
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount}. New balance: ${self.__balance}")
        else:
            print("Insufficient balance or invalid amount.")

    def _calculate_interest(self):  # PROTECTED method
        return self.__balance * 0.05

    def __get_balance(self):        # PRIVATE method
        return self.__balance

    def display_info(self):         # PUBLIC method
        print(f"Account Holder: {self.name}")
        print(f"Branch: {self._branch}")
        print(f"Balance: ${self.__balance}")


# Creating account
account = BankAccount("John Doe", 1000)

# 1. PUBLIC member access
print("\n--- PUBLIC ACCESS ---")
print(f"Account name (public): {account.name}")
account.deposit(500)
account.withdraw(200)

# 2. PROTECTED member access
print("\n--- PROTECTED ACCESS ---")
print(f"Branch (protected): {account._branch}")
interest = account._calculate_interest()
print(f"Interest (protected method): ${interest:.2f}")

# 3. PRIVATE member access (via name mangling)
print("\n--- PRIVATE ACCESS (Name Mangling) ---")
print(f"Private balance via name mangling: {account._BankAccount__balance}")
private_balance = account._BankAccount__get_balance()
print(f"Private method returned: {private_balance}")

# 4. Attempting direct private access (will fail)
print("\n--- DIRECT PRIVATE ACCESS (Will Fail) ---")
try:
    print(account.__balance)
except AttributeError as e:
    print(f"Error: {e}")

# 5. Display all info using public method
print("\n--- COMPLETE ACCOUNT INFO ---")
account.display_info()
````

## Comparison Table: Specifiers/Modifiers

| Feature | PUBLIC | PROTECTED | PRIVATE |
|---------|--------|-----------|---------|
| **Symbol** | No underscore | Single underscore `_` | Double underscore `__` |
| **Syntax** | `self.data` | `self._data` | `self.__data` |
| **Scope** | Anywhere (cross-package) | Within same package | Within same class only |
| **Access** | `obj.data` | `obj._data` | `obj._ClassName__data` |
| **Inheritance** | Available in child | Available in child | Not directly available |
| **Data Protection** | No protection | Limited protection | Strong protection |

## Advantages of Encapsulation

1. **Data Hiding** - Prevents direct access to data
2. **Increased Security** - Protects data from unauthorized access
3. **Flexibility** - Can change internal implementation without affecting other code
4. **Reusability** - Encapsulated code can be reused
5. **Testing** - Easier to test and debug
6. **Maintainability** - Code is easier to maintain and modify

## Key Takeaways

1. **Encapsulation** = Binding data + methods into a single unit (class)

2. **Three Specifiers**:
   - **PUBLIC** (no underscore) - Accessible everywhere
   - **PROTECTED** (`_`) - Accessible within package
   - **PRIVATE** (`__`) - Accessible only within class

3. **Private members** can still be accessed via **name mangling**:
   - `object._ClassName__private_member`


4. **Encapsulation provides**:
   - Data protection
   - Controlled access
   - Validation logic
   - Flexibility to change implementation
