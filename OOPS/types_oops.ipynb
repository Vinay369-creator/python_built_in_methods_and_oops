# Five Major Pillars of Object-Oriented Programming (OOPS)

Object-Oriented Programming contains **5 major pillars** that form the foundation of OOP design and implementation.

## The Five Pillars

| # | Pillar | Description |
|---|--------|-------------|
| 1 | **ENCAPSULATION** | Binding properties and methods into a single roof |
| 2 | **AGGREGATION** | Accessing another class object into the current class object |
| 3 | **INHERITANCE** | Accessing another class into the current class |
| 4 | **ABSTRACTION** | Hiding basic functionality from the user |
| 5 | **POLYMORPHISM** | Behaving differently from one object to another object |

---

## 1. ENCAPSULATION

**Definition:** Encapsulation is the process of binding properties (attributes) and methods (functions) into a single unit or roof (class).

### Key Concepts:
- Bundling data and methods that operate on that data within a single unit
- Restricting direct access to some of an object's components
- Data hiding and protection



---

## 2. AGGREGATION

**Definition:** Aggregation is the process of accessing another class's object into the current class object.

### Key Concepts:
- Represents "HAS-A" relationship
- One class contains an object of another class
- The contained object can exist independently



---

## 3. INHERITANCE

**Definition:** Inheritance is the process of accessing another class into the current class.

### Key Concepts:
- Represents "IS-A" relationship
- Child class acquires properties and methods of parent class
- Promotes code reusability

### Types of Inheritance in Python:
- Single Inheritance
- Multiple Inheritance
- Multilevel Inheritance
- Hierarchical Inheritance
- Hybrid Inheritance



---

## 4. ABSTRACTION

**Definition:** Abstraction is the process of hiding the basic functionality (implementation details) from the user.

### Key Concepts:
- Showing only essential features to the user
- Hiding internal implementation details
- Achieved through abstract classes and interfaces



---

## 5. POLYMORPHISM

**Definition:** Polymorphism means having many forms. It behaves differently from one object to another object.

### Key Concepts:
- Same method name but different implementations
- Method Overloading (not directly supported in Python)
- Method Overriding
- Duck Typing



---

## Advantages of OOPS

| Advantage | Description |
|-----------|-------------|
| **Code Reusability** | By using OOPs, code reusability is high (through inheritance) |
| **Low Code Redundancy** | Code redundancy is low (DRY principle) |
| **Easy Debugging** | Debugging process becomes easier due to modular structure |
| **Low Complexity** | Complexity is low as problems are broken into objects |
| **Code Modularity** | Code modularity can be achieved through classes and objects |
| **Data Protection** | Data protection is achieved through encapsulation |
| **Data Hiding** | Data is hidden through abstraction |
| **Maintainability** | Easier to maintain and update code |
| **Scalability** | Easy to scale applications |
| **Real-world Modeling** | Better representation of real-world entities |

---

| Pillar | Definition | Key Concept | Real-world Example |
|--------|------------|-------------|-------------------|
| **Encapsulation** | Binding properties and methods into a single roof | Data hiding, protection | Bank account with private balance |
| **Aggregation** | Accessing another class object into current class object | HAS-A relationship | Employee HAS-A Address |
| **Inheritance** | Accessing another class into current class | IS-A relationship, code reuse | Car IS-A Vehicle |
| **Abstraction** | Hiding basic functionality from the user | Interface vs Implementation | Abstract Shape class |
| **Polymorphism** | Behaving differently from one object to another | Many forms, method overriding | Different animals making different sounds |
